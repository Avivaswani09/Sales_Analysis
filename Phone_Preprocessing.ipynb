```python
import re

# Function to extract RAM
def extract_ram(ram_string):
    if isinstance(ram_string, str):
        match = re.search(r'(\d+)\s*GB RAM', ram_string)
        return int(match.group(1)) if match else None
    return None

# Function to extract inbuilt storage
def extract_inbuilt_storage(ram_string):
    if isinstance(ram_string, str):
        match = re.search(r'(\d+)\s*GB inbuilt', ram_string)
        return int(match.group(1)) if match else None
    return None

# Apply the functions to create new columns
phone_data['ram'] = phone_data['ram'].apply(extract_ram)
phone_data['storage'] = phone_data['ram'].apply(extract_inbuilt_storage)
```

```python
# Function to extract battery life from the 'battery' column
def extract_battery_life(battery_string):
    match = re.search(r'(\d+)\s*mAh', battery_string)
    return int(match.group(1)) if match else None

# Apply the function to create a new column 'battery_life'
phone_data['battery_life'] = phone_data['battery'].apply(extract_battery_life)

# Display the first few rows to verify the changes
phone_data[['battery', 'battery_life']].head()
```

```python
# Function to extract charging speed from the 'battery' column
def extract_charging_speed(battery_string):
    match = re.search(r'(\d+)\s*W', battery_string)
    return int(match.group(1)) if match else None

# Apply the function to create a new column 'charging_speed'
phone_data['charging_speed'] = phone_data['battery'].apply(extract_charging_speed)

# Display the first few rows to verify the changes
phone_data[['battery', 'battery_life', 'charging_speed']].head()
```

```python
import re

# Ensure the 'display' column values are strings
phone_data['display'] = phone_data['display'].astype(str)

# Function to extract screen size from the 'display' column
def extract_screen_size(display_string):
    match = re.search(r'(\d+\.\d+)\s*inches', display_string)
    return float(match.group(1)) if match else None

# Function to extract display resolution from the 'display' column
def extract_display_resolution(display_string):
    match = re.search(r'(\d+)\s*x\s*(\d+)', display_string)
    return (int(match.group(1)), int(match.group(2))) if match else (None, None)

# Apply the functions to create new columns
phone_data['screen_size'] = phone_data['display'].apply(extract_screen_size)
phone_data['resolution_width'], phone_data['resolution_height'] = zip(*phone_data['display'].apply(extract_display_resolution))

# Display the first few rows to verify the changes
phone_data[['display', 'screen_size', 'resolution_width', 'resolution_height']].head()
```

```python
import re

# Function to count the number of rear cameras, subtracting one if a front camera is mentioned
def count_rear_cameras(camera_string):
    if isinstance(camera_string, str):
        rear_camera_count = camera_string.count('MP')
        if 'Front' in camera_string:
            rear_camera_count -= 1
        return rear_camera_count
    return None

# Function to extract the highest resolution of the rear cameras
def extract_primary_camera_resolution(camera_string):
    if isinstance(camera_string, str):
        matches = re.findall(r'(\d+)\s*MP', camera_string)
        if matches:
            return max(int(match) for match in matches)
    return None

# Apply the function to create a new column 'primary_camera_resolution'
phone_data['primary_camera_resolution'] = phone_data['camera'].apply(extract_primary_camera_resolution)

# Apply the function to update the 'num_rear_cameras' column
phone_data['num_rear_cameras'] = phone_data['camera'].apply(count_rear_cameras)

# Display the first few rows to verify the changes
phone_data[['camera', 'num_rear_cameras', 'primary_camera_resolution']].head()
```

```python
# Function to extract the resolution of the front camera
def extract_front_camera_resolution(camera_string):
    if isinstance(camera_string, str):
        match = re.search(r'(\d+)\s*MP Front', camera_string)
        return int(match.group(1)) if match else None
    return None

# Apply the function to create a new column 'front_camera_resolution'
phone_data['front_camera_resolution'] = phone_data['camera'].apply(extract_front_camera_resolution)

# Display the first few rows to verify the changes
phone_data[['camera', 'front_camera_resolution']].head()
```
